```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import random as r
```

```python
#Read Data From CSV file
df=pd.read_csv('https://raw.githubusercontent.com/DamanVirdiNITK/BestAddPredictor/5aba96be4bb3cf356e21ebc1689af771c36a3c15/Ads_CTR_Optimisation.csv')
df.head()
```

```python
#Implement Thompsons Sampling
N=10000
d=10
ads_select=[]
no_of_rewards1=[0]*d
no_of_rewards0=[0]*d
total_rewards=0

for i in range (0,N):
  ad=0
  max_random=0
  for j in range(0,d):
    random_beta=r.betavariate(no_of_rewards1[j]+1,no_of_rewards0[j]+1)
    if random_beta>max_random:
      max_random=random_beta
      ad=j
  ads_select.append(ad)
  reward=df.values[i,ad]
  if reward==1:
    no_of_rewards1[ad]+=1
  else:
    no_of_rewards0[ad]+=1
  total_rewards+=reward
```

```python
#Visualize Result Using MatPlotLib Library
plt.hist(ads_select)
plt.title("BEST ADS PREDICTOR USING THOMPSONS SAMPLING")
plt.xlabel("ADS")
plt.ylabel("TIMES ADS SELECTED")
```
